```python
# Libraries
# if matplotlib is not working, try to install it with pip3 install matplotlib
from matplotlib import pyplot as plt
from typing import List, Tuple
```

```python
years: List[int] = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp: List[int] = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# add a title
plt.title("Nominal GDP")

# add a label to the y-axis
plt.ylabel("Billions of $")

# add a label to the x-axis
plt.xlabel("Years")

# show the plot
plt.show()
```

```python
movies: List[str] = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars: List[int] = [5, 11, 3, 8, 10]

plt.bar(range(len(movies)), num_oscars)

plt.title("My Favorite Movies")
plt.ylabel("# of Academy Awards")

plt.xticks(range(len(movies)), movies)

plt.show()
```
